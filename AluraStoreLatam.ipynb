
#1. Análisis de facturación



```python
import pandas as pd
# URLs de las tiendas
urls = {
  'Tienda 1': "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
  'Tienda 2': "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
  'Tienda 3': "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
  'Tienda 4': "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

# tienda = pd.read_csv(url)
# tienda2 = pd.read_csv(url2)
# tienda3 = pd.read_csv(url3)
# tienda4 = pd.read_csv(url4)

# cargar los datos en un diccionario de DataFrames
tiendas = {nombre: pd.read_csv(url) for nombre, url in urls.items()}

# calcular la facturación total por tienda
facturacion_total = {nombre: df['Precio'].sum() for nombre, df in tiendas.items()}

# mostrar los resultados
print('\nFacturación total por tienda:')
for nombre, total in facturacion_total.items():
  print(f'{nombre}: {total:,.2f}')
```

```python
# gráfico de barras
import matplotlib.pyplot as plt

# crea gráfico
plt.figure(figsize=(8, 6))
plt.bar(facturacion_total.keys(), facturacion_total.values(), color='skyblue')
plt.xlabel('Tiendas')
plt.ylabel('Facturación Total')
plt.title('Facturación total por tienda')
plt.grid(axis = 'y', linestyle = '--', alpha = 0.9)

# agregar etiquetas a las barras
for i, v in enumerate(facturacion_total.values()):
  plt.text(i, v, f'{v:,.1f}', ha = 'center', va = 'bottom', fontsize = 10)

# mostrar gráfico
plt.show()
```

# 2. Ventas por categoría

```python
# calcular ventas por categorías en cada tienda
ventas_por_categoria = {
    nombre: df.groupby('Categoría del Producto')['Precio'].sum()
    for nombre, df in tiendas.items()
}

# mostrar resultados
print('\nVentas por categoría: ')
for nombre, ventas in ventas_por_categoria.items():
  print(f'\n{nombre}:')
  print(ventas.apply(lambda x: f'{x:,.2f}').to_string())
```

```python
import matplotlib.pyplot as plt

# obtener las 4 categorías más vendidas en general
ventas_totales_por_categoria = pd.concat([df.groupby('Categoría del Producto')['Precio'].sum() for df in tiendas.values()])
ventas_totales_por_categoria = ventas_totales_por_categoria.groupby(ventas_totales_por_categoria.index).sum()
top_categorias = ventas_totales_por_categoria.sort_values(ascending = False).head(4).index

# crear la proporción de ventas por tienda dentro de cada una de las 4 categorías más vendidas
for categoria in top_categorias:
  ventas_por_tienda = {nombre: df[df['Categoría del Producto'] == categoria]['Precio'].sum()
                        for nombre, df in tiendas.items()}
# personalización de los gráficos
  plt.figure(figsize = (5, 5))
  plt.pie(ventas_por_tienda.values(), labels = ventas_por_tienda.keys(),
          autopct = '%1.1f%%', startangle = 90, colors = plt.cm.Paired.colors)
  plt.title(f'Proporción de tiendas dentro de la categoría {categoria}')
  plt.tight_layout()
  plt.show()
```

# 3. Calificación promedio de la tienda


```python
# calcular la calificación promedio por tienda
calificacion_promedio = {nombre: df['Calificación'].mean() for nombre, df in tiendas.items()}

# mostrar resultados
print('\n Calificación promedio por tienda:')
for nombre, promedio in calificacion_promedio.items():
  print(f'{nombre}: {promedio:.2f}')
```

```python
import matplotlib.pyplot as plt

# crear gráfico de línea
plt.figure(figsize = (8, 5))
plt.plot(list(calificacion_promedio.keys()), list(calificacion_promedio.values()), marker = 'o', color = 'royalblue')

# Personalización del gráfico
plt.title('Calificación promedio por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Calificación promedio')
plt.ylim(min(calificacion_promedio.values()) - 0.5, max(calificacion_promedio.values()) + 0.5)
plt.grid(True, linestyle = '--', alpha = 0.7)
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Iterar sobre cada tienda para contar la frecuencia de los productos vendidos
for tienda, url in urls.items():
    df = pd.read_csv(url)

    # Contar la frecuencia de cada producto en la columna "Producto"
    conteo_productos = df["Producto"].value_counts()

    print(f"\n{tienda}:")

    # Mostrar los productos más vendidos
    print("\nProductos más vendidos:")
    max_vendidos = conteo_productos[conteo_productos == conteo_productos.max()]
    for producto, cantidad in max_vendidos.items():
        print(f"{producto}: {cantidad} veces")

    # Mostrar los productos menos vendidos
    print("\nProductos menos vendidos:")
    min_vendidos = conteo_productos[conteo_productos == conteo_productos.min()]
    for producto, cantidad in min_vendidos.items():
        print(f"{producto}: {cantidad} unidades")
```

# 5. Envío promedio por tienda

```python
# calcular el costo de envío promedio por tienda
envio_promedio = {nombre: df['Costo de envío'].mean() for nombre, df in tiendas.items()}

# mostrar resultados
print('\nCosto de envío promedio por tienda:')
for nombre, promedio in envio_promedio.items():
  print(f'{nombre}: {promedio:,.2f}')
```
