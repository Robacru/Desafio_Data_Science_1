
#1. Análisis de facturación



```python
import pandas as pd
# URLs de las tiendas
urls = {
  'Tienda 1': "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
  'Tienda 2': "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
  'Tienda 3': "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
  'Tienda 4': "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# cargar los datos en un diccionario de DataFrames
tiendas = {nombre: pd.read_csv(url) for nombre, url in urls.items()}

# calcular la facturación total por tienda
facturacion_total = {nombre: df['Precio'].sum() for nombre, df in tiendas.items()}

# mostrar los resultados
print('\nFacturación total por tienda:')
for nombre, total in facturacion_total.items():
  print(f'{nombre}: {total:,.2f}')
```

# 2. Ventas por categoría

```python
# calcular ventas por categorías en cada tienda
ventas_por_categoria = {
    nombre: df.groupby('Categoría del Producto')['Precio'].sum()
    for nombre, df in tiendas.items()
}

# mostrar resultados
print('\nVentas por categoría: ')
for nombre, ventas in ventas_por_categoria.items():
  print(f'\n{nombre}:')
  print(ventas.apply(lambda x: f'{x:,.2f}').to_string())
```

# 3. Calificación promedio de la tienda


```python
# calcular la calificación promedio por tienda
calificacion_promedio = {nombre: df['Calificación'].mean() for nombre, df in tiendas.items()}

# mostrar resultados
print('\n Calificación promedio por tienda:')
for nombre, promedio in calificacion_promedio.items():
  print(f'{nombre}: {promedio:.2f}')
```

# 4. Productos más y menos vendidos

```python
# calcular ventas por productos en cada tienda
ventas_por_producto = {
    nombre: df.groupby('Producto')['Precio'].sum().sort_values(ascending = False)
    for nombre, df in tiendas.items()
}

# mostrar resultados
print('\nProductos más y menos vendidos por tienda:')
for nombre, ventas in ventas_por_producto.items():
  print(f'\n{nombre} - Productos ordenados por ventas:')
  print(ventas.apply(lambda x: f'{x:,.2f}').to_string())
```

# 5. Envío promedio por tienda

```python
# calcular el costo de envío promedio por tienda
envio_promedio = {nombre: df['Costo de envío'].mean() for nombre, df in tiendas.items()}

# mostrar resultados
print('\nCosto de envío promedio por tienda:')
for nombre, promedio in envio_promedio.items():
  print(f'{nombre}: {promedio:,.2f}')
```
